<b>Comsol</b>

A silicon substrate was choosen with dielectric constant 11.7 and coplanar waveguide structure was designed with appropriate dimension and perfect electric conductor boundary condition. Two Lumped port was designed which couples resonator with capacitive coupling.
During simulation the length of coplanar waveguide which acts as a resonator was varied using parametric sweep to study the change of resonator frequency with the variation in length.
Also, the coupling capacitance was varied by changing the gap between plates, to study its effect on quality factor and line width of the resonator. 

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
import math
```

<b> parametric sweep of gap to change capacitive coupling </b>

```python

filename = 'E:\IISc\quantum_lab\lab\cpw\cpwgapvsf.csv'
#data = np.loadtxt(filename, delimiter=',', skiprows=1, dtype=str)
opens = pd.read_csv(filename) 
```

```python
opens
```

```python
opens.plot.scatter(x='cap_gap(um)', y=['ResonantFrequency(GHz)'], figsize=(10,5), grid=True) 
opens.plot.scatter(x='cap_gap(um)', y=['Qualityfactor(1)'], figsize=(10,5), grid=True) 
```

The captured data from comsol contains manhy eigen frequency out of which one signifies the actual resonant frequency. Therefore, the scatter plot of resonant frequency vs Cap gap & Quality factor vs Cap gap show multiple points.
As the length of resonator is kept fixed so no variation in resonant frequency was observed but the quality factor corresponding to eigen frequency which is actual resonant frequency got varied.

```python
y=list(opens['Qualityfactor(1)'])
x=list(opens['cap_gap(um)'])
l=list(opens['linewidth(Hz)'])
y1=[]
x1=[]
l1=[]
k=len(y)
for i in range(0,k):
    if y[i]>13000000: #to remove unwanted eigenfrequencies
        y1.append(y[i])
        x1.append(x[i])
        l1.append(l[i])

print(y1)
plt.plot(x1,y1)
plt.xlabel("capacitance gap(um)")
plt.ylabel("Quality Factor")
plt.figure()
plt.plot(x1,l1)
plt.xlabel("capacitance gap(um)")
plt.ylabel("Line width(Hz)")
plt.show()
```

<b> Result 1 </b>

It can be observed from the above graph that with the increase in capacitive gap, capacitive coupling decreases and hence quality factor of the resonator decreases and line width increases. As the perfect electric conductor boundary condition is implemented for CPW, so the effect of line width is mainly coming from external line width Ke1 and Ke2
